```
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
import numpy as np
iris = load_iris()
data = iris["data"][:, :4]
setosa = (iris["target"]==0).astype(np.int)
versi = (iris["target"]==1).astype(np.int)
virgi = (iris["target"]==2).astype(np.int)
train_data, test_data, train_labels, test_labels = train_test_split(data, setosa, test_size=0.3, random_state=101)
```

```
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
log_reg = LogisticRegression()
log_reg.fit(train_data, train_labels)
print(train_data.shape)
print(test_data.shape)
y_proba = log_reg.predict_proba(test_data)
pred_setosa = log_reg.predict(test_data)

plt.plot(test_data, y_proba[:, 1], "g-", label="Iris-Setosa")
plt.plot(test_data, y_proba[:, 0], "b--", label="Not Iris-Setosa")
print(accuracy_score(test_labels, pred_setosa, normalize=True)*100, "%")
```

```
train_data, test_data, train_labels, test_labels = train_test_split(data, versi, test_size=0.3, random_state=101)
log_reg.fit(train_data, train_labels)
y_proba = log_reg.predict_proba(test_data)
pred_versi = log_reg.predict(test_data)

plt.plot(test_data, y_proba[:, 1], "g-", label="Iris-Versicolor")
plt.plot(test_data, y_proba[:, 0], "b--", label="Not Iris-Versicolor")
print(accuracy_score(test_labels, pred_versi, normalize=True)*100, "%")
```

```
train_data, test_data, train_labels, test_labels = train_test_split(data, virgi, test_size=0.3, random_state=101)
log_reg.fit(train_data, train_labels)
y_proba = log_reg.predict_proba(test_data)
pred_virgi = log_reg.predict(test_data)

plt.plot(test_data, y_proba[:, 1], "g-", label="Iris-Virginica")
plt.plot(test_data, y_proba[:, 0], "b--", label="Not Iris-Verginica")
print(accuracy_score(test_labels, pred_virgi, normalize=True)*100, "%")
```
